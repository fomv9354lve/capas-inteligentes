# Run a CAPAS claim yourself — no install of trust required

CAPAS does **not** determine truth. It decides whether supplied evidence **licenses** a claim, deterministically. Run the cells: you get the same verdict and the same audit hash we publish. That is the whole point — reproduce it on infra we don't control.

```python
# Install CAPAS (Binder clones the repo; on Colab this pulls it from GitHub)
try:
    import capas_sdk
except Exception:
    import subprocess, sys
    subprocess.run([sys.executable,'-m','pip','install','-q','git+https://github.com/fomv9354lve/capas-inteligentes'], check=False)
    import capas_sdk
```

## 1. Claim drift: association evidence does NOT license a causal claim

```python
out = capas_sdk.gate('statistical_confidence', {'p_value':0.03,'alpha':0.05,'effect_direction_confirmed':False}, 'X reduces defects by 35%')
print(out['verdict'], '-', out.get('why',''))
```

## 2. Pharma statistical admissibility (the beachhead): claims significance the p-value doesn't support

```python
import capas_pharma
r = capas_pharma.gate_pharma_stat_claim({'p_value':0.08,'alpha':0.05,'asserts_significant':True})
print(r['verdict'], '-', r['why'])
```

## 3. Deterministic: same input -> same verdict (run it twice)

```python
ev = {'abs_error':0.0,'tolerance':1e-3}
a = capas_sdk.gate('exact_model_solution', ev, 'demo'); b = capas_sdk.gate('exact_model_solution', ev, 'demo')
print('same verdict:', a['verdict']==b['verdict'], '->', a['verdict'])
```
